```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mu,sd=5,2
a,b= mu-3*sd, mu+3*sd
x=np.linspace(a,b,100)
y=norm.pdf(x,loc=mu,scale=sd)
plt.plot(x,y,'.g')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Calculate the mean and standard deviation of the data

df = pd.read_excel(r"C:\Users\BAJRANG\Downloads\preprocessed-type-0.xlsx", sheet_name='Sheet1')

#data = df['tumor_size']

data = df['p_id']
mu = data.mean()

print("mean=",mu)

sigma = data.std()

print("deviation=",sigma)


x = np.linspace(mu - 3*sigma, mu + 3*sigma, 1000)

# Calculate the probability density function (PDF) for the normal distribution
y = norm.pdf(x, mu, sigma)




# Set the axis labels and title
plt.xlabel('Data')
plt.ylabel('Probability Density')
plt.title('Normal Distribution on Breast Cancer Tumor Size')

# Plot the normal distribution curve
plt.plot(x, y)
plt.show()


```
